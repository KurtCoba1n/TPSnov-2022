```python
sub_files= sorted(glob.glob(r'C:\Users\Shreedeep\Downloads\tps\subs/*.csv'))
```

```python
scores=[float(file_name.split('\\')[-1][:-4]) for file_name in sub_files]
scores.sort()
scores
```

```python
best = '0.6222863195.csv'
else_cols = [f for f in data.columns if f != best]
drop_cols = []
for col, score in tqdm(zip(else_cols, scores)):
    logloss = score
    r, _ = spearmanr(data[best].values, data[col].values)
    if r > 0.98 or logloss > 0.680:
        drop_cols.append(col)
```

```python
train.drop(drop_cols, axis=1, inplace=True)
print("train.shape:", train.shape)
```

```python
from lightgbm import LGBMClassifier
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import log_loss
```

```python
import umap
```

```python
embedding=umap.UMAP(n_neighbors=50,min_dist=0.5,metric='manhattan').fit_transform(train)
```

```python
preds = cross_val_predict(LGBMClassifier(), embedding, labels["label"], cv=5, method="predict_proba")
print("Manhattan score:", log_loss(labels["label"], preds))
colors_dict = {
        0: 'tab:cyan',
        1: 'tab:red',
}
fig, ax = plt.subplots(figsize=(8,8))
for color_key in colors_dict.keys():
    indexs = labels["label"] == color_key
    temp_embedding = embedding[indexs, :]
    ax.scatter(temp_embedding[:, 0], temp_embedding[:, 1], 
                c=colors_dict[color_key], 
                edgecolor='none', 
                alpha=0.80,
                label=color_key,
                s=10)
ax.legend(bbox_to_anchor=(1, 1), fontsize="x-large", markerscale=2.)
ax.set_ylabel("UMAP_1")
ax.set_xlabel("UMAP_0")
ax.set_title('UMAP - Manhattan', fontsize=18);
```
